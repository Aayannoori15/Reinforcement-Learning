```python
import gymnasium as gym 
import numpy as np 
import random 
```

```python
#sample env 
env=gym.make('CliffWalking-v1')
```

## SARSA

```python
gamma=0.9
alpha=0.5
epsilon=0.1
episodes=500
```

```python
Q=np.zeros((48,4))
```

```python
def epsilon_greedy(state):
    if random.random()<epsilon:
        return env.action_space.sample()
    else:
        return np.argmax(Q[state])
```

```python
for episode in range(episodes):
    env=gym.make('CliffWalking-v1')
    done=False 
    state,_=env.reset()
    action=epsilon_greedy(state)
    total_reward=0
    episode_len=0
    while not done:
        next_state,reward,terminate,truncate,_=env.step(action)
        done=terminate or truncate 
        next_action=epsilon_greedy(next_state)
        Q[state,action]+=alpha*(reward+gamma*(Q[next_state,next_action]-Q[state,action]))
        state=next_state
        action=next_action
        total_reward+=reward 
        episode_len+=1
    print(f'total reward in this episode {episode+1} is {total_reward}')
```

## now lets test our agent

```python
env=gym.make('CliffWalking-v1')
done=False 
state,_=env.reset()
total_reward=0
episode_len=0
while not done:
    action=np.argmax(Q[state])
    next_state,reward,terminate,truncate,_=env.step(action)
    done=terminate or truncate 
    state=next_state
    total_reward+=reward 
    episode_len+=1
print(f'total reward in this episode is {total_reward} and episode length is {episode_len}')
```
